# **Pre- preprocessing**

```python
df = pd.read_csv('/content/gdrive/MyDrive/Term1/Special Topics on Software Engineering/Project/Final Trimmed Dataset/Qualitative Data Master Old.csv')
df
```

```python
df[df.isnull().any(axis = 1)]
```

```python
df.dropna(axis = 0, how = 'any', inplace = True)
```

```python
df.isnull().sum()
```

```python
df['QuestionUserID'] = df['QuestionUserID'].astype(int)
df['AnswerUserID'] = df['AnswerUserID'].astype(int)
df['CommentUserID'] = df['CommentUserID'].astype(int)
```

```python
def ByWho(row):
  AUI = row['AnswerUserID']
  CUI = row['CommentUserID']
  QUI = row['QuestionUserID']
  if AUI == CUI:
    return 'answerer'
  elif QUI == CUI:
    return 'asker'
  elif (QUI != CUI) or (QUI != AUI) or (AUI != CUI):
    return 'outsider'
  else: 
    return 'commenter'

df['ByWho'] = df.apply(lambda x: ByWho(x), axis = 1)
```

```python
def keyword(row):
  comment = row['CommentText']
  if 'deprecated' in comment:
    return 'deprecated'
  elif 'obsolete' in comment:
    return 'obsolete'
  elif 'out of date' in comment:
    return 'out of date'
  elif 'outdated' in comment:
    return 'outdated'
  elif 'old-fashioned' in comment:
    return 'old-fashioned'
  elif 'old' in comment:
    return 'old'
  elif 'ancient' in comment:
    return 'ancient'
  elif 'behind the times' in comment:
    return 'behind the times'
  elif 'archaic' in comment:
    return 'archaic'
  else:
    return 'outworn'

df['Keywords'] = df.apply(lambda x: keyword(x), axis = 1)
```

```python
df = df.drop(['PostTypeId'], axis = 1)
df
```
